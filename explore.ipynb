```python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import gaussian_kde
import plotly.io as pio

# Set renderer
pio.renderers.default = 'notebook'  # or 'iframe' or 'browser'

# Filter and clean data
data = df_train[
    (df_train["Arrival Delay in Minutes"] >= 0) &
    (df_train["Arrival Delay in Minutes"] <= 120)
]["Arrival Delay in Minutes"].dropna()

# KDE calculation
kde = gaussian_kde(data)
x_vals = np.linspace(0, 120, 1000)
y_vals = kde(x_vals)

# Create figure
fig = go.Figure()

# Base KDE line (outline)
fig.add_trace(go.Scatter(
    x=x_vals,
    y=y_vals,
    mode='lines',
    line=dict(color='green', width=3),
    name='KDE Density',
    fill=None
))

# Define cumulative thresholds and corresponding fill colors
thresholds = [5, 10, 15, 30, 60, 120]
colors = [
    'rgba(0, 100, 0, 0.15)',
    'rgba(0, 128, 0, 0.2)',
    'rgba(34, 139, 34, 0.25)',
    'rgba(60, 179, 113, 0.3)',
    'rgba(46, 139, 87, 0.35)',
    'rgba(0, 100, 0, 0.4)'
]

# Add cumulative shaded areas
for threshold, color in zip(thresholds, colors):
    x_section = x_vals[x_vals <= threshold]
    y_section = kde(x_section)

    fig.add_trace(go.Scatter(
        x=np.concatenate([[0], x_section, [threshold]]),
        y=np.concatenate([[0], y_section, [0]]),
        fill='toself',
        fillcolor=color,
        line=dict(color='rgba(0,0,0,0)'),
        showlegend=False,
        name=f"≤ {threshold} min"
    ))

# Final layout
fig.update_layout(
    title="Cumulative Density Plot of Arrival Delays (Up to 120 Minutes)",
    xaxis_title="Arrival Delay (minutes)",
    yaxis_title="Density",
    plot_bgcolor="lightgray",
    paper_bgcolor="lightgray",
    font=dict(color="black")
)

fig.show()
```

```python
# Cleaned delay data
delays = df_train["Arrival Delay in Minutes"].dropna()
delays = delays[(delays >= 0) & (delays <= 120)]

# Sort and compute cumulative distribution
sorted_delays = np.sort(delays)
cdf = np.arange(1, len(sorted_delays)+1) / len(sorted_delays)

# Plot
plt.figure(figsize=(8,5))
plt.plot(sorted_delays, cdf, color='darkgreen')
plt.title("Cumulative Probability of Arrival Delays")
plt.xlabel("Delay Time (minutes)")
plt.ylabel("Cumulative Probability")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(delays, fill=True, color='seagreen')
plt.xscale('log')
plt.title("KDE with Log-Scaled Delay Time")
plt.xlabel("Delay Time (minutes, log scale)")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
# Assume cost = delay²
delays_under_120 = delays[delays <= 120]
costs = delays_under_120 ** 2

plt.figure(figsize=(8,5))
plt.plot(delays_under_120, costs, '.', alpha=0.3, color='crimson')
plt.title("Nonlinear Cost of Delay (Delay²)")
plt.xlabel("Delay Time (minutes)")
plt.ylabel("Estimated Cost")
plt.grid(True)
plt.show()
```

```python
pio.renderers.default = 'notebook'

# Prepare data: 0 to 120 mins
data = df_train[
    (df_train["Arrival Delay in Minutes"] >= 0) &
    (df_train["Arrival Delay in Minutes"] <= 600)
]["Arrival Delay in Minutes"].dropna()

# KDE setup
kde = gaussian_kde(data)
x_vals = np.linspace(0, 600, 500)
density = kde(x_vals)

# Define cost function
cost = x_vals**2  # Quadratic cost

# Compute weighted density
weighted_density = density * cost

# Create plot
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=x_vals,
    y=weighted_density,
    mode='lines',
    line=dict(color='crimson', width=3),
    fill='tozeroy',
    fillcolor='rgba(220, 20, 60, 0.3)',
    name='Weighted Density'
))

fig.update_layout(
    title="Weighted Density of Arrival Delays (KDE × Delay² Cost)",
    xaxis_title="Arrival Delay (minutes)",
    yaxis_title="Weighted Impact (Density × Cost)",
    plot_bgcolor="lightgray",
    paper_bgcolor="lightgray",
    font=dict(color="black")
)

fig.show()
```

```python
pio.renderers.default = 'notebook'

# Prepare data: 0 to 600 mins
data = df_train[
    (df_train["Arrival Delay in Minutes"] >= 0) &
    (df_train["Arrival Delay in Minutes"] <= 1000)
]["Arrival Delay in Minutes"].dropna()

# KDE setup
kde = gaussian_kde(data)
x_vals = np.linspace(0, 1000, 500)
density = kde(x_vals)

# Define cost function
cost = x_vals**2  

# Compute weighted density
weighted_density = density * cost

# Apply smoothing with a Univariate Spline
spline = UnivariateSpline(x_vals, weighted_density, s=75) 
smoothed_weighted_density = spline(x_vals)

# Create plot
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=x_vals,
    y=smoothed_weighted_density,
    mode='lines',
    line=dict(color='crimson', width=3),
    fill='tozeroy',
    fillcolor='rgba(220, 20, 60, 0.3)',
    name='Smoothed Weighted Density'
))

fig.update_layout(
    title="Smoothed Weighted Density of Arrival Delays (KDE × Delay² Cost)",
    xaxis_title="Arrival Delay (minutes)",
    yaxis_title="Weighted Impact (Density × Cost)",
    plot_bgcolor="lightgray",
    paper_bgcolor="lightgray",
    font=dict(color="black")
)

fig.show()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from scipy.stats import gaussian_kde

# Load your real dataset here
data = df_train["Flight Distance"].dropna()

# KDE for original
kde_orig = gaussian_kde(data, bw_method=0.3)
x_vals_orig = np.linspace(data.min(), data.max(), 500)
y_vals_orig = kde_orig(x_vals_orig)

# KDE for log-transformed
log_data = np.log(data)
kde_log = gaussian_kde(log_data, bw_method=0.3)
x_vals_log = np.linspace(log_data.min(), log_data.max(), 500)
y_vals_log = kde_log(x_vals_log)

# Create subplots
fig = make_subplots(rows=1, cols=2)

# Plot original KDE
fig.add_trace(go.Scatter(
    x=x_vals_orig, y=y_vals_orig,
    mode='lines',
    line=dict(color='green', width=4),
    fill='tozeroy',
    fillcolor='rgba(0, 128, 0, 0.2)'
), row=1, col=1)

# Plot log KDE
fig.add_trace(go.Scatter(
    x=x_vals_log, y=y_vals_log,
    mode='lines',
    line=dict(color='green', width=4),
    fill='tozeroy',
    fillcolor='rgba(0, 128, 0, 0.2)'
), row=1, col=2)

# Layout
fig.update_layout(
    plot_bgcolor="white",
    paper_bgcolor="white",
    font=dict(color="black"),
    showlegend=False
)
fig.update_xaxes(title_text="Flight Distance", gridcolor='rgba(0,0,0,0.1)', row=1, col=1)
fig.update_yaxes(title_text="Density", gridcolor='rgba(0,0,0,0.1)', row=1, col=1)
fig.update_xaxes(title_text="Log(Flight Distance)", gridcolor='rgba(0,0,0,0.1)', row=1, col=2)
fig.update_yaxes(gridcolor='rgba(0,0,0,0.1)', row=1, col=2)

fig.show()
```
